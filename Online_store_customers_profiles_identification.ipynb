#### Обработка дат

```python
# Перевод столбца с датой в формат даты
df['date'] =  pd.to_datetime(df['date'], format='%Y%m%d%H')
df.sample(3)
```

```python
# Перевод наименований продуктов в нижний регистр
df['product'] = df['product'].str.lower()
df.sample(3)
```

#### Детализация даты заказа

```python
# Добаление столбцов с детализацией даты
df['order_date'] = df['date'].astype('datetime64[D]')
df['hour'] = df.date.dt.hour
df['day'] = (df.date.dt.dayofweek) + 1
df['week'] = df.date.dt.week
df['month'] = df.date.dt.month
df['year'] = df.date.dt.year
df['month'] = df['date'].astype('datetime64[M]')
```

```python
# Расчет дат первого визита
first_order = df.groupby(['customer_id'])['order_date'].min()
first_order.name = 'first_order'
first_order.sample(3).to_frame().reset_index()
```

```python
# Добавление столбца с датой первого заказа
df = df.merge(first_order, on='customer_id')
df.sample(2)
```

**Обработана дата и добавлены столбцы детализирующие дату заказа по дням недели, недели, месяцу и году для анализа динамики метрик во времени, дата первого заказа - для расчета retention покупателей.**

#### Обработка частичных дубликатов

```python
# Проверка на присутствие уникальных id заказов в разных датах
order_date_dublicated = df.groupby('order_id').agg({'order_date':'nunique'}).reset_index()
order_date_dublicated = order_date_dublicated.query('order_date > 1').sort_values('order_date', ascending=False)
order_date_dublicated
```

**Выявленно 190 закзов с уникальными id но с разными датами**

```python
df.query('order_id == 71148')
```

**Заказы с одним id но разными датами, id покупателя тоже не различается как и наименование товара. Возможно какая-то ошибка в логе, такие заказы - дубли в пределах дня допустимы но в разных датах искажают динамику во временном периоде, поэтому возможно оставить заказы только с первой датой.**

```python
# Выделение id дублированных заказов в лист
dublicated_orders = order_date_dublicated['order_id'].to_list()
```

```python
# Фильтр по дублированным id заказов, выделение заказов с датой заказа старше даты первого закза
dub_set = df.query('order_id in @dublicated_orders')
drop_set = dub_set.query('order_date > first_order')
```

#### Расчет customer retention rate

```python
# Создание столбцов месяца первого заказа и месяца заказа, форматирование их в формат первого дня месяца
df['first_order_month'] = df['first_order'].astype('datetime64[M]')
df.sample(2)
```

```python
# Расчет Lifetime в рамках когорты
df['lifetime'] = (df['month'] - df['first_order_month']) / np.timedelta64(1, 'M')
df['lifetime'] = df['lifetime'].round().astype('int')
df.sample(2)
```

```python
# Группировка количества покупателей по месяцу первого заказа и lifetime
cohort_orders = df.groupby(['first_order_month', 'lifetime']).agg({'customer_id': 'nunique'}).reset_index()
# Подсчет изначального количества покупателей в когорте
сohort_customers = cohort_orders[cohort_orders['lifetime'] == 0][['first_order_month', 'customer_id']]
сohort_customers.columns = ['first_order_month', 'сohort_customer']
сohort_customers.sample(3)
```

```python
# Cлияние когорты, количества покупателей, lifetime  c когортой и количеством изначальных покупателей
cohort_orders = cohort_orders.merge(сohort_customers, on='first_order_month')
cohort_orders.sample(3)
```

```python
# Расчет Retention Rate
cohort_orders['retention'] = cohort_orders['customer_id'] / cohort_orders['сohort_customer']
cohort_orders['first_order_month'] = cohort_orders['first_order_month'].dt.strftime('%Y-%m') 
cohort_orders.sample(3)
```

```python
# Группировка значений Retention Rate по корготам и месяцам lifetime
customer_retention_pivot = cohort_orders.pivot_table(
    index = 'first_order_month', 
columns = 'lifetime', values = 'retention', aggfunc = 'sum')
customer_retention_pivot = customer_retention_pivot*100
```

```python
# Тепловая карта Retention Rate
plt.figure(figsize=(12,9))
sns.heatmap(customer_retention_pivot, linewidths=.4, annot=True, fmt='2.2f', vmin=0, vmax=5) 
plt.title('Динамика Retention Rate')
plt.xlabel('lifetime')
plt.ylabel('когорты')
plt.show()
```

```python
# Расчет средних значений по месяцу lifetime
avg_rr = pd.DataFrame(customer_retention_pivot.mean()).reset_index()
avg_rr.columns = ['lifetime_month', 'average_rr_pct']
avg_rr.style.hide_index().format({'average_rr_pct' : "{:,.2f}"})
```

```python
# Добавление среднего Retention Rate по второму месяцу в сводную таблицу метрик
arr = avg_rr.loc[1,:]
arr = pd.DataFrame(
    {'metric' : 
        'Средний % Retention во второй мес.',
     'value' : [arr[1]]})
product_metrics = product_metrics.append(arr).reset_index(drop=True)
product_metrics
```

## <span style='background :lightblue' >Проверка гипотез</span>

### Проверка различий в кластерах покупателей по выручке

<br>**$Ho$**  - ***средняя выручка в 0 и 1 кластере не различается***
<br>**$H1$**  - ***средняя выручка в 0 и 1 кластере различается***

<br>**$Ho$**  - ***средняя выручка в 1 и 2 кластере не различается***
<br>**$H1$**  - ***средняя выручка в 1 и 2 кластере различается***

<br>**$Ho$**  - ***средняя выручка в 2 и 3 кластере не различается***
<br>**$H1$**  - ***средняя выручка в 2 и 3 кластере различается***

<br>**$Ho$**  - ***средняя выручка 3 и 4 кластере не различается***
<br>**$H1$**  - ***средняя выручка 3 и 4 кластере различается***

```python
# Cбор номеров кластеров в лист
clusters_list = clusters_result['cluster_km'].to_list()
```

```python
# Лист для сбора p-value
plist1 = []
```

```python
# Применение тетса Манна-Уитни для попарного сравнения кластеров по выручке
for i in clusters_list:
    if i != (len(clusters_list)-1):
          
        cluster0 = clusters_data[clusters_data['cluster_km']==i]['sales']
        cluster1 = clusters_data[clusters_data['cluster_km']==i+1]['sales']
        print('Дисперсия'+' '+ str(i) + ' '+ 'кластера: {} '.format(np.var(cluster0, ddof=1)))
        print('Дисперсия'+' '+ str(i+1) + ' '+ 'кластера: {} '.format(np.var(cluster1, ddof=1)))
        print("Соотношение дисперсий: {} ".format((np.var(cluster1, ddof=1))/(np.var(cluster0, ddof=1))))
        
        alpha = 0.05 # критический уровень статистической значимости 
        # Поскольку есть 0 и альтернативная гипотеза, нужно включить параметр 'two=sided'
        results = stats.mannwhitneyu(cluster0, cluster1, alternative='two-sided')
        # Вывод значения p-value
        print('p-значение: {:.20f}'.format(results.pvalue))

        if (results.pvalue < alpha):
    
            print('\033[1m' + "Отвергаем нулевую гипотезу" + '\033[0m')
            plist1.append(results.pvalue) # Сбор p-value
            print(' ')
        else:
            print('\033[1m' + "Не получилось отвергнуть нулевую гипотезу" + '\033[0m')
            print(' ')
    else:
        break
```

```python
# Расчет поправки на множественность теста
reject, p_value_corrected, sidak_corr, bonf_corr = multipletests(plist1, alpha=0.05, method='holm-sidak')
reject.tolist()
```

```python
# Расчет скорректированных p-value
p_value_corrected.tolist()
```

**Все кластеры показали статистически значимые различия по выручке, поправка на множественность также не изменила вывод**

```python
# Формирование выборок по выручке в будние и выходные.
sales_weekdays = df[df['weekdays']==1]['sales']
sales_weekends = df[df['weekends']==1]['sales']
```

```python
print("Дисперсия объём продаж в будни: {} ".format(np.var(sales_weekdays, ddof=1)))
print("Дисперсия объём продаж в выходные: {} ".format(np.var(sales_weekends, ddof=1)))
print("Соотношение дисперсий: {} ".format((np.var(sales_weekdays, ddof=1))/(np.var(sales_weekends, ddof=1))))

print("Стандартное отклонение объём продаж в будни: {} ".format(np.std(sales_weekdays, ddof=1)))
print("Стандартное отклонение объём продаж в выходные: {} ".format(np.std(sales_weekends, ddof=1)))
```

```python
# Проверка различий в выручке в будние и выходные с T тестом
alpha = 0.05

results = st.ttest_ind(sales_weekdays, sales_weekends, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('\033[1m' + "Отвергаем нулевую гипотезу" + '\033[0m')
else:
    print('\033[1m' + "Не получилось отвергнуть нулевую гипотезу" + '\033[0m')
```

**Принимая во внимание что дисперсия двух выборок различается и результат теста можно сказать что средняя выручка в будние дни и в выходные различается**
